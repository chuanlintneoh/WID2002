```python
# Question 1

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

L = np.arange(0, 300, 0.1)
K = np.arange(0, 300, 0.1)
L, K = np.meshgrid(L, K)
P = L**0.75 * K**0.25

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(L, K, P, cmap=cm.coolwarm)
ax.set_xlabel('K')
ax.set_ylabel('L')
ax.set_zlabel('P')
plt.tight_layout()
plt.show()
```

```python
# Question 2

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

x = np.arange(-2, 2, 0.1)
y = np.arange(-4, 4, 0.1)
x, y = np.meshgrid(x, y)
h = 4*x**2 + y**2

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x, y, h, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('h')

plt.figure()
plt.contour(x, y, h)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Question 8 b (Illustration 1)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

x = np.arange(-98, 102, 0.1)
y = np.arange(-104, 96, 0.1)
x, y = np.meshgrid(x, y)
f = x**3 * y + 12*x**2 - 8*y

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x, y, f, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')
ax.scatter(2, -4, 48, color='k', label='Saddle point')
ax.legend()

plt.figure()
plt.contour(x, y, f)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```

```python
# Saddle Point Illustration Example

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)
x, y = np.meshgrid(x, y)
f = x**2 + y**3

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x, y, f, cmap=cm.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f')
ax.scatter(0, 0, 0, color='k', label='Saddle point')
ax.legend()

plt.figure()
plt.contour(x, y, f)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```
