```python
# Question 3
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**2
# or any function

def f_prime(x):
  return 2*x
# derivate of above function

def linear_approximation(a, x):
  return f(a) + f_prime(a)*(x-a)

x_values = np.linspace(-10, 10, 100)

y_values = f(x_values)

linear_approx_values = linear_approximation(2, x_values)

plt.plot(x_values, y_values, label='f(x) = x^2')
plt.plot(x_values, linear_approx_values, label='Linear Approximation at x = 2', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Approximation vs. Original Function')
plt.legend()
plt.grid()
plt.show()
```

```python
# Question 6
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return x**2 + 1

def f2(x):
  return x**2 - 4*x + 5

x1 = np.arange(-4, 1, 0.01)
x2 = np.arange(1.01, 6, 0.01)

plt.plot(x1, f1(x1), label='f1(x) = x^2 + 1')
plt.plot(x2, f2(x2), label='f2(x) = x^2 - 4x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
# Question 7
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**3 - 4*x**2 + x + 2

def f_prime(x):
  return 3*x**2 - 8*x + 1

x1 = np.arange(-10, 10, 0.01)
x2 = np.arange(-10, 10, 0.01)

# Plot the first function
plt.figure()
plt.plot(x1, f_prime(x1), label='f\'(x) = 3x^2 -8x + 1', color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Graph of f\'(x) = 3x^2 -8x + 1')
plt.show()

# Plot the second function
plt.figure()
plt.plot(x2, f(x2), label='f(x) = x^3 - 4x^2 + x + 2', color='orange')
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Graph of f(x) = x^3 - 4x^2 + x + 2')
plt.show()
```
