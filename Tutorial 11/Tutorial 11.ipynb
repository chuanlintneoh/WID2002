```python
# Question 1

import numpy as np
import matplotlib.pyplot as plt

def a(x):
    return np.sin(2*x)

def b(x):
    return np.sin(x)*np.cos(x)

def c(x):
    return np.tan(x)

def d(x):
    return x**2

def e(z):
    return z**3

def f(x):
    return np.abs(x)

a_domain = np.linspace(-np.pi, np.pi, 100)
b_domain = np.linspace(-np.pi, np.pi, 100)
c_domain = np.linspace(-np.pi/2, np.pi/2, 1000)
domain = np.linspace(-10, 10, 100)

fig, axes = plt.subplots(6, 3, figsize=(15, 18))

functions = [a, b, c, d, e, f]
domains = [a_domain, b_domain, c_domain, domain, domain, domain]
titles = ['$f(x) = sin(2x)$', '$f(x) = sin(x)cos(x)$', '$f(x) = tan(x)$',
          '$f(x) = x^2$', '$f(x) = x^3$', '$f(x) = |x|$']

for i, (func, domain, title) in enumerate(zip(functions, domains, titles)):
    axes[i, 0].plot(domain, func(domain), color='black')
    axes[i, 0].set_title(f'Original: {title}')
    axes[i, 0].grid()
    axes[i, 0].set_xlabel('x')
    axes[i, 0].set_ylabel('f(x)')
    axes[i, 0].axhline(0, color='black', lw=1)
    axes[i, 0].axvline(0, color='black', lw=1)
    
    axes[i, 1].plot(domain, func(-domain), color='green')
    axes[i, 1].set_title(f'Even Symmetry Test: {title}')
    axes[i, 1].grid()
    axes[i, 1].set_xlabel('x')
    axes[i, 1].set_ylabel('f(x)')
    axes[i, 1].axhline(0, color='black', lw=1)
    axes[i, 1].axvline(0, color='black', lw=1)
    
    axes[i, 2].plot(-domain, -func(domain), color='blue')
    axes[i, 2].set_title(f'Odd Symmetry Test: {title}')
    axes[i, 2].grid()
    axes[i, 2].set_xlabel('x')
    axes[i, 2].set_ylabel('f(x)')
    axes[i, 2].axhline(0, color='black', lw=1)
    axes[i, 2].axvline(0, color='black', lw=1)
    
    if func == c:
        axes[i, 0].set_ylim(-10, 10)
        axes[i, 1].set_ylim(-10, 10)
        axes[i, 2].set_ylim(-10, 10)

plt.tight_layout()
plt.show()
```

```python
# Question 2

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x * np.sqrt(x**2 + np.cos(x))

x_values = np.linspace(-2*np.pi, 2*np.pi, 100)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].plot(x_values, f(x_values), color='black')
axes[0].set_title('Original: $f(x) = x\sqrt{x^2 + cos(x)}$')
axes[0].grid()
axes[0].set_xlabel('x')
axes[0].set_ylabel('f(x)')
axes[0].axhline(0, color='black', lw=1)
axes[0].axvline(0, color='black', lw=1)

axes[1].plot(x_values, f(-x_values), color='green')
axes[1].set_title('Even Symmetry Test: $f(x) = x\sqrt{x^2 + cos(x)}$')
axes[1].grid()
axes[1].set_xlabel('x')
axes[1].set_ylabel('f(x)')
axes[1].axhline(0, color='black', lw=1)
axes[1].axvline(0, color='black', lw=1)

axes[2].plot(-x_values, -f(x_values), color='blue')
axes[2].set_title('Odd Symmetry Test: $f(x) = x\sqrt{x^2 + cos(x)}$')
axes[2].grid()
axes[2].set_xlabel('x')
axes[2].set_ylabel('f(x)')
axes[2].axhline(0, color='black', lw=1)
axes[2].axvline(0, color='black', lw=1)

plt.tight_layout()
plt.show()
```

```python
# Question 3

def a(m, n, x):
  return np.sin(2*np.pi*m*x) * np.sin(2*np.pi*n*x)

def b(m, n, x):
  return np.cos(2*np.pi*m*x) * np.cos(2*np.pi*n*x)

x_values = np.linspace(-np.pi, np.pi, 100)

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

functions = [a, b]
domains = [x_values, x_values]
titles = ['$f(x) = sin(2\pi mx)sin(2\pi nx)$', '$f(x) = cos(2\pi mx)cos(2\pi nx)$']

# m = n
# assume a = m = n = 3
a = 3

for i, (func, domain, title) in enumerate(zip(functions, domains, titles)):
  axes[i, 0].plot(domain, func(a, a, domain), color='black')
  axes[i, 0].set_title(f'Original: {title}')
  axes[i, 0].grid()
  axes[i, 0].set_xlabel('x')
  axes[i, 0].set_ylabel('f(x)')
  axes[i, 0].axhline(0, color='black', lw=1)
  axes[i, 0].axvline(0, color='black', lw=1)
  
  axes[i, 1].plot(domain, func(a, a, -domain), color='green')
  axes[i, 1].set_title(f'Even Symmetry Test: {title}')
  axes[i, 1].grid()
  axes[i, 1].set_xlabel('x')
  axes[i, 1].set_ylabel('f(x)')
  axes[i, 1].axhline(0, color='black', lw=1)
  axes[i, 1].axvline(0, color='black', lw=1)
  
  axes[i, 2].plot(-domain, -func(a, a, domain), color='blue')
  axes[i, 2].set_title(f'Odd Symmetry Test: {title}')
  axes[i, 2].grid()
  axes[i, 2].set_xlabel('x')
  axes[i, 2].set_ylabel('f(x)')
  axes[i, 2].axhline(0, color='black', lw=1)
  axes[i, 2].axvline(0, color='black', lw=1)

plt.tight_layout()
plt.show()
```

```python
# Question 4

import numpy as np
import matplotlib.pyplot as plt

def a(x):
  return np.where((x >= 0) & (x <= np.pi), np.pi, 0)
def b(x):
  return np.exp(x)
def c(x):
  return x/2
def lecture_example(x):
  return x

domain_a = np.linspace(-np.pi, np.pi, 100)
domain_b = np.linspace(-np.pi, np.pi, 100)
domain_c = np.linspace(0, 2*np.pi, 100)
domain_example = np.linspace(-np.pi, np.pi, 100)

def fourier_series(a0, an_func, bn_func, x, n):
  result = a0
  for num_terms in range(1, n + 1):
    result += an_func(num_terms) * np.cos(num_terms*x) + bn_func(num_terms) * np.sin(num_terms*x)
  return result

def an_b(n):
  return (1/np.pi)*(
    (np.exp(np.pi) - np.exp(-np.pi))*(((-1)**n)/(n**2)) +
    (np.exp(-np.pi) - np.exp(np.pi))*(((-1)**n)/((n**2)*(n**2 + 1)))
    )

def bn_a(n):
  return (1-(-1)**n)/n
def bn_b(n):
  return (1/np.pi)*(
    ((np.exp(np.pi)*(-1)**(n+1))/n) -
    ((np.exp(np.pi)*(-1)**(n+1))/(n*(n**2 + 1))) -
    ((np.exp(-np.pi)*(-1)**(n+1))/n) +
    ((np.exp(-np.pi)*(-1)**(n+1))/(n*(n**2 + 1)))
    )
def bn_c(n):
  return -1/n
def bn_example(n):
  return (2/n) * (-1)**(n+1)

fig, axes = plt.subplots(4, 2, figsize=(10, 20))

functions = [a, b, c, lecture_example]
domains = [domain_a, domain_b, domain_c, domain_example]
titles = [
  '$f(x) = 0, \pi$',
  '$f(x) = e^x$',
  '$f(x) = \\frac{x}{2}$',
  '$f(x) = x$'
]
fourier_series_titles = [
  '$f(x) = \\frac{\pi}{2} + \sum_{n=1}^{\infty} \\frac{2}{n} sin(nx)$',
  # '$f(x) = \\frac{\exp(\pi) - \exp(-\pi)}{2\pi} + \sum_{n=1}^{\infty} \left( \\frac{1}{\pi} \left( (\\exp(\pi) - \\exp(-\pi)) \\frac{(-1)^n}{n^2} + (\\exp(-\pi) - \\exp(\pi)) \\frac{(-1)^n}{n^2(n^2 + 1)} \\right) \\cos(nx) + \\frac{1}{\pi} \left( \\frac{\\exp(\pi)(-1)^{n+1}}{n} - \\frac{\\exp(\pi)(-1)^{n+1}}{n(n^2 + 1)} - \\frac{\\exp(-\pi)(-1)^{n+1}}{n} + \\frac{\\exp(-\pi)(-1)^{n+1}}{n(n^2 + 1)} \\right) \\sin(nx) \\right)$',
  'hidden...',
  '$f(x) = \\frac{\pi}{2} + \sum_{n=1}^{\infty} -\\frac{1}{n} sin(nx)$',
  '$f(x) = \sum_{n=1}^{\infty} (\\frac{2}{k} (-1)^{n+1} sin(nx))$'
]
a0_constants = [
  np.pi/2,
  (np.exp(np.pi) - np.exp(-np.pi)) / (2*np.pi),
  np.pi/2,
  0
]
an_funcs = [
  lambda n: 0,
  an_b,
  lambda n: 0,
  lambda n: 0
]
bn_funcs = [
  bn_a,
  bn_b,
  bn_c,
  bn_example
]
num_terms = 100

for i, (func, domain, title, fourier_series_title, a0_constant, an_func, bn_func) in enumerate(zip(functions, domains, titles, fourier_series_titles, a0_constants, an_funcs, bn_funcs)):
  axes[i, 0].plot(domain, func(domain), color='black')
  axes[i, 0].set_title(title)
  axes[i, 0].grid()
  axes[i, 0].set_xlabel('x')
  axes[i, 0].set_ylabel('f(x)')
  axes[i, 0].axhline(0, color='black', lw=1)
  axes[i, 0].axvline(0, color='black', lw=1)
  
  axes[i, 1].plot(domain, fourier_series(a0_constant, an_func, bn_func, domain, num_terms), color='red')
  axes[i, 1].set_title(f'Fourier Series Approximation: {fourier_series_title}')
  axes[i, 1].grid()
  axes[i, 1].set_xlabel('x')
  axes[i, 1].set_ylabel('f(x)')
  axes[i, 1].axhline(0, color='black', lw=1)
  axes[i, 1].axvline(0, color='black', lw=1)

plt.tight_layout()
plt.show()
```

```python
# Question 4 (animated)
# Copy and paste the code below into a python file to view the graph animation, as it is not supported in Jupyter Notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

def a(x):
  return np.where((x >= 0) & (x <= np.pi), np.pi, 0)
def b(x):
  return np.exp(x)
def c(x):
  return x/2
def lecture_example(x):
  return x

domain_a = np.linspace(-np.pi, np.pi, 100)
domain_b = np.linspace(-np.pi, np.pi, 100)
domain_c = np.linspace(0, 2*np.pi, 100)
domain_example = np.linspace(-np.pi, np.pi, 100)

def fourier_series(a0, an_func, bn_func, x, n):
  result = a0
  for num_terms in range(1, n + 1):
    result += an_func(num_terms) * np.cos(num_terms*x) + bn_func(num_terms) * np.sin(num_terms*x)
  return result

def an_b(n):
  return (1/np.pi)*(
    (np.exp(np.pi) - np.exp(-np.pi))*(((-1)**n)/(n**2)) +
    (np.exp(-np.pi) - np.exp(np.pi))*(((-1)**n)/((n**2)*(n**2 + 1)))
    )

def bn_a(n):
  return (1-(-1)**n)/n
def bn_b(n):
  return (1/np.pi)*(
    ((np.exp(np.pi)*(-1)**(n+1))/n) -
    ((np.exp(np.pi)*(-1)**(n+1))/(n*(n**2 + 1))) -
    ((np.exp(-np.pi)*(-1)**(n+1))/n) +
    ((np.exp(-np.pi)*(-1)**(n+1))/(n*(n**2 + 1)))
    )
def bn_c(n):
  return -1/n
def bn_example(n):
  return (2/n) * (-1)**(n+1)

animations = [
  {
    'func': a,
    'domain': domain_a,
    'title': '$f(x) = 0, \pi$',
    'fourier_series_title': '$f(x) = \\frac{\pi}{2} + \sum_{n=1}^{\infty} \\frac{2}{n} sin(nx)$',
    'a0_constant': np.pi/2,
    'an_func': lambda n: 0,
    'bn_func': bn_a
  },
  {
    'func': b,
    'domain': domain_b,
    'title': '$f(x) = e^x$',
    # 'fourier_series_title': '$f(x) = \\frac{\exp(\pi) - \exp(-\pi)}{2\pi} + \sum_{n=1}^{\infty} \left( \\frac{1}{\pi} \left( (\\exp(\pi) - \\exp(-\pi)) \\frac{(-1)^n}{n^2} + (\\exp(-\pi) - \\exp(\pi)) \\frac{(-1)^n}{n^2(n^2 + 1)} \\right) \\cos(nx) + \\frac{1}{\pi} \left( \\frac{\\exp(\pi)(-1)^{n+1}}{n} - \\frac{\\exp(\pi)(-1)^{n+1}}{n(n^2 + 1)} - \\frac{\\exp(-\pi)(-1)^{n+1}}{n} + \\frac{\\exp(-\pi)(-1)^{n+1}}{n(n^2 + 1)} \\right) \\sin(nx) \\right)$',
    'fourier_series_title': 'hidden...',
    'a0_constant': (np.exp(np.pi) - np.exp(-np.pi)) / (2*np.pi),
    'an_func': an_b,
    'bn_func': bn_b
  },
  {
    'func': c,
    'domain': domain_c,
    'title': '$f(x) = \\frac{x}{2}$',
    'fourier_series_title': '$f(x) = \\frac{\pi}{2} + \sum_{n=1}^{\infty} -\\frac{1}{n} sin(nx)$',
    'a0_constant': np.pi/2,
    'an_func': lambda n: 0,
    'bn_func': bn_c
  },
  {
    'func': lecture_example,
    'domain': domain_example,
    'title': '$f(x) = x$',
    'fourier_series_title': '$f(x) = \sum_{n=1}^{\infty} (\\frac{2}{k} (-1)^{n+1} sin(nx))$',
    'a0_constant': 0,
    'an_func': lambda n: 0,
    'bn_func': bn_example
  }
]
num_terms = 1000

def animate_graph(animation_choice):
  selected_animation = animations[animation_choice]
  
  func = selected_animation['func']
  domain = selected_animation['domain']
  title = selected_animation['title']
  fourier_series_title = selected_animation['fourier_series_title']
  a0_constant = selected_animation['a0_constant']
  an_func = selected_animation['an_func']
  bn_func = selected_animation['bn_func']
  
  fig, ax = plt.subplots(figsize=(10, 5))
  
  def update(frame):
    ax.cla()
    ax.plot(domain, func(domain), color='black', label=title)
    ax.plot(domain, fourier_series(a0_constant, an_func, bn_func, domain, frame), color='red', label='Fourier Approximation')
    ax.set_title(f'Fourier Series Approximation: {fourier_series_title}')
    ax.grid()
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.axhline(0, color='black', lw=1)
    ax.axvline(0, color='black', lw=1)
    ax.legend()
  
  anim = FuncAnimation(fig, update, frames=range(1, num_terms + 1), interval=200)
  
  plt.tight_layout()
  plt.show()

user_choice = int(input("Enter the number corresponding to the animation (0: a, 1: b, 2: c, 3: lecture_example): "))
animate_graph(user_choice)
```
