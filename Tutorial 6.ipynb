```python
# Question 1

import sympy as sp

x = sp.symbols('x')
f = sp.exp(x) * sp.sin(x)

integral = sp.integrate(f, x)

print("Integral of e^x * sin(x) dx:")
print(integral)
```

```python
# Question 2

import sympy as sp

x, y = sp.symbols('x y')

integrand = x**2 * y

result = sp.integrate(integrand, (y, 1, 2), (x, 0, 3))

print("The result of the double integral is:", result)
```

```python
# Question 3

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x_values = np.linspace(-1, 2, 300)

y1_values = np.exp(x_values)  # y = e^x
y2_values = x_values           # y = x

plt.plot(x_values, y1_values, label='$y = e^x$')
plt.plot(x_values, y2_values, label='$y = x$')

plt.fill_between(x_values, y1_values, y2_values, where=(x_values >= 0) & (x_values <= 1), color='lightblue', alpha=0.5)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Region Bounded by $y = e^x$ and $y = x$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()

x = sp.Symbol('x')
f1 = sp.exp(x)  # y = e^x
f2 = x          # y = x

area = sp.integrate(f1 - f2, (x, 0, 1))
print("Area of the region bounded = ", area.evalf())
```

```python
# Question 4

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x_values = np.linspace(-0.5, 1.5, 400)

y1_values = x_values**2
y2_values = x_values*2 - x_values**2

plt.plot(x_values, y1_values, label='$y = x^2$')
plt.plot(x_values, y2_values, label='$y = 2x - x^2$')

x = sp.Symbol('x')
f1 = x**2
f2 = 2*x - x**2
x_intersections = sp.solve(sp.Eq(f1,f2), x)

print("Intersection x values = ", x_intersections)

plt.fill_between(x_values, y1_values, y2_values, where=(x_values >= x_intersections[0]) & (x_values <= x_intersections[1]), color='lightblue', alpha=0.5)

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Region Bounded by $y = x^2$ and $y = 2x - x^2$')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

plt.show()

x = sp.Symbol('x')
f1 = x**2
f2 = 2*x - x**2

area = sp.integrate(f2 - f1, (x, x_intersections[0], x_intersections[1]))
print("Area of the region bounded = ", area.evalf())
```

```python
# Question 9

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return (x**3)**0.5

x = np.arange(0, 5, 0.01)
x_arc = np.arange(1, 4, 0.01)

plt.figure()
plt.plot(x, f(x), label='$y^2 = x^3$', color='blue')
plt.plot(1, 1, 'ro', label='(1, 1)')
plt.plot(4, 8, 'ro', label='(4, 8)')
plt.plot(x_arc, f(x_arc), color='red')
plt.ylim(0, 9)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Graph of $y^2 = x^3$')
plt.show()
```

```python
# Question 10

import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return np.where((x >= 0) & (x <= 10), 0.006*x*(10-x), 0)
  
x = np.arange(-1, 11, 0.01)

plt.figure()
plt.plot(x, f(x), label='f(x)', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.title('Graph of f(x)')
plt.show()
```
