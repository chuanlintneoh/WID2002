```python
# Question 2
import matplotlib.pyplot as plt
import numpy as np

x1 = np.arange(0.1, 2.99, 0.01)
x2 = np.arange(3.01, 5, 0.01)
def f(x):
  return 2*x/(x-3)

plt.plot(x1, f(x1))
plt.plot(x2, f(x2))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# Question 7 b
import matplotlib.pyplot as plt
import numpy as np

x1 = np.arange(-np.pi/2, np.pi/2, 0.01)
x2 = np.arange(np.pi/2 + 0.01, 3*np.pi/2, 0.01)
def f(x):
  return np.tan(x)

plt.plot(x1, f(x1))
plt.plot(x2, f(x2))
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# Question 12
import sympy as sp

x = sp.symbols('x')

f_a = 1/x**2
f_b = x**(2/3)
f_c = 4 * sp.log(x) * sp.sin(x**3)
f_d = 1 / (1 + sp.exp(-x))

f_a_diff = sp.diff(f_a, x)
f_b_diff = sp.diff(f_b, x)
f_c_diff = sp.diff(f_c, x)
f_d_diff = sp.diff(f_d, x)

print("a. f'(x) =", f_a_diff)
print("b. f'(x) =", f_b_diff)
print("c. f'(x) =", f_c_diff)
print("d. f'(x) =", f_d_diff)
```

```python
# Question 16
import sympy as sp

x, y = sp.symbols('x y')

equation = x**2 + y**3 - 4

derivative_eqn = sp.diff(equation, x)

dy_dx = sp.solve(derivative_eqn, sp.diff(y, x))

print("dy/dx =", dy_dx)
```
