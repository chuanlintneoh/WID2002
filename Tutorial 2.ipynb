```python
# Question 2

import matplotlib.pyplot as plt
import numpy as np

w = np.arange(0.1, 10, 0.1)
C = 20 * w**2 + 180/w
plt.plot(w, C)
plt.grid()
plt.xlabel('width (m)')
plt.ylabel('cost ($)')
```
