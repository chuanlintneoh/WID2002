```python
# Question 6

import matplotlib.pyplot as plt
import numpy as np

dataset_1 = [19, 24, 12, 19, 18, 24, 8, 5, 9, 20, 13, 11, 1, 12, 11, 10, 22, 21, 7, 16, 15, 15, 26, 16, 1, 13, 21, 21, 20, 19]
dataset_2 = [17, 24, 21, 22, 26, 22, 19, 21, 23, 11, 19, 14, 23, 25, 26, 15, 17, 26, 21, 18, 19, 21, 24, 18, 16, 20, 21, 20, 23, 33]
dataset_3 = [56, 52, 13, 34, 33, 18, 44, 41, 48, 75, 24, 19, 35, 27, 46, 62, 71, 24, 66, 94, 40, 18, 15, 39, 53, 23, 41, 78, 15, 35]

def plot_histogram(data, title):
  plt.hist(data, bins=10, color='blue', edgecolor='black')
  plt.title(title)
  plt.xlabel('Value')
  plt.ylabel('Frequency')
  plt.show()

plot_histogram(dataset_1, 'Dataset 1')
plot_histogram(dataset_2, 'Dataset 2')
plot_histogram(dataset_3, 'Dataset 3')

print(f"Dataset 1 - Mean: {np.mean(dataset_1)}, Median: {np.median(dataset_1)}, Skew Direction: {'Right-Skewed' if np.mean(dataset_1) > np.median(dataset_1) else 'Left-Skewed' if np.median(dataset_1) > np.mean(dataset_1) else 'Symmetric'}")
print(f"Dataset 2 - Mean: {np.mean(dataset_2)}, Median: {np.median(dataset_2)}, Skew Direction: {'Right-Skewed' if np.mean(dataset_2) > np.median(dataset_2) else 'Left-Skewed' if np.median(dataset_2) > np.mean(dataset_2) else 'Symmetric'}")
print(f"Dataset 3 - Mean: {np.mean(dataset_3)}, Median: {np.median(dataset_3)}, Skew Direction: {'Right-Skewed' if np.mean(dataset_3) > np.median(dataset_3) else 'Left-Skewed' if np.median(dataset_3) > np.mean(dataset_3) else 'Symmetric'}")
```

```python
# Question 7

import numpy as np

data = [43, 37, 50, 51, 58, 105, 52, 45, 45, 10]

print('Before Removing Outliers: ')
print(data)
print(f"Mean: {np.mean(data)}")
print(f"Median: {np.median(data)}")
print(f"Quartiles: {np.percentile(data, [25, 50, 75])}")
print(f"Standard Deviation: {np.std(data)}")

print()
IQR = np.percentile(data, [25, 75])[1] - np.percentile(data, [25, 75])[0]
outlier_bottom_range = np.percentile(data, [25, 75])[0] - 1.5 * IQR
outlier_top_range = np.percentile(data, [25, 75])[1] + 1.5 * IQR
print(f"Outlier Range: {outlier_bottom_range} - {outlier_top_range}")
print()

for i in data:
  if i < outlier_bottom_range or i > outlier_top_range:
    data.remove(i)

print('After Removing Outliers: ')
print(data)
print(f"Mean: {np.mean(data)}")
print(f"Median: {np.median(data)}")
print(f"Quartiles: {np.percentile(data, [25, 50, 75])}")
print(f"Standard Deviation: {np.std(data)}")
```
